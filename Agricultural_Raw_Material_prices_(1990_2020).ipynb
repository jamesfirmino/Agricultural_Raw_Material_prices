```python
#Import libraries
import numpy as np
import pandas as pd
from datetime import datetime

#Import file
df = pd.read_csv('agricultural_raw_material.csv')
```

```python
df.info
#Check Null values
df.isnull().sum()
```

```python
#Clean missing, incorrect, and invalid data
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)
#Drop row with NaN values
df = df.dropna()
#Check if all NaN values are fixed
df.isnull().sum()
#Convert data type to float
lst = ["Coarse wool Price",
       "Coarse wool price % Change",
       "Copra Price",
       "Copra price % Change",
       "Cotton Price",
       "Cotton price % Change",
       "Fine wool Price",
       "Fine wool price % Change",
       "Hard log Price",
       "Hard log price % Change",
       "Hard sawnwood Price",
       "Hard sawnwood price % Change",
       "Hide Price",
       "Hide price % change",
       "Plywood Price",
       "Plywood price % Change",
       "Rubber Price",
       "Rubber price % Change",
       "Softlog Price",
       "Softlog price % Change",
       "Soft sawnwood Price",
       "Soft sawnwood price % Change",
       "Wood pulp Price",
       "Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
df['Month'] = df['Month'].astype(str)  # Convert Month column to string type
df['Month'] = pd.to_datetime(df['Month'].str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')
df.head()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'

#raw data list
raw_data = ["Coarse wool Price",
       "Copra Price",
       "Cotton Price",
       "Fine wool Price",
       "Hard log Price",
       "Hard sawnwood Price",
       "Hide Price",
       "Plywood Price",
       "Rubber Price",
       "Softlog Price",
       "Soft sawnwood Price",
       "Wood pulp Price"]

#Getting the correlation matrix
corr_matrix = df[raw_data].corr()

#Image plot size
fig = plt.figure(figsize = (12, 9))

#Masking the upper triangle since matrix is symetric
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, vmax = .8, mask=mask, square =True, annot=True)
plt.show()
```

The heat map shows the correlation between the higher raw material value and the higher correlated chance of being two related raw materials.

```python
fig = plt.figure(figsize = (30, 15))
changelist = [ "Coarse wool price % Change",
               "Copra price % Change",
               "Cotton price % Change",
               "Fine wool price % Change",
               "Hard log price % Change",
               "Hard sawnwood price % Change",
               "Hide price % change",
               "Plywood price % Change",
               "Rubber price % Change",
               "Softlog price % Change",
               "Soft sawnwood price % Change",
               "Wood pulp price % Change"]

corrMatrix = df[changelist].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

From this graph, we can see that there is no relationship between the percentage variation of raw material prices.

1.   A negative value implies that two variables are negatively correlated.
2.   Zero implies no relationship.
3.   The higher the positive value, the higher the chance of price correlation and their percentage change graphs.

```python
axes = df[["Coarse wool Price", "Coarse wool price % Change"]].plot(figsize=(11, 9), subplots=True, linewidth=1)
```

```python
changelist = ["Coarse wool price % Change","Copra price % Change","Cotton price % Change","Fine wool price % Change","Hard log price % Change","Hard sawnwood price % Change", "Hide price % change", "Plywood price % Change", "Rubber price % Change", "Softlog price % Change", "Soft sawnwood price % Change", "Wood pulp price % Change"]
for index in range(len(changelist)):
  plt.figure(figsize=(12,12))
  df[changelist[index]].hist(figsize=(11, 9), linewidth=1)
  plt.xlabel('% change')
  plt.ylabel('Count')
  plt.legend(changelist[index:], loc='upper center', bbox_to_anchor=(1.2, 1))
```

It is observed that the majority of raw materials have an ideal frequent percentage change below 5%.

```python
plt.figure(figsize=(10, 10))
raw_material_list = ["Coarse wool Price","Copra Price", "Cotton Price", "Fine wool Price", "Hard log Price", "Hard sawnwood Price", "Hide Price", "Plywood Price", "Rubber Price", "Softlog Price", "Soft sawnwood Price", "Wood pulp Price"]
for index in range(len(raw_material_list)):
  plt.subplot(4,3, index+1)
  plt.subplots_adjust( hspace=1, wspace=0.5)
  plt.title(raw_material_list[index])
  plt.plot(df[raw_material_list[index]])
  plt.xticks(rotation=90)
plt.suptitle("Raw material price comparison")
```

It is observed that cotton and rubber have the lowest prices.

```python
plt.figure(figsize=(10, 10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.title("Raw material price comparison")
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Price', 'Rubber Price'],loc='upper center',bbox_to_anchor=(1.2,1))
```

Cotton has been the lowest-priced raw material in recent years.
